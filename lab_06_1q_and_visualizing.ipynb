## q-sphere

To the right of the Probabilities chart, is the <i>q-sphere</i>.


Screenshot from 2024-05-24 11-17-19.png
The q-sphere represents the state of a system of one or more qubits by associating each computational basis state with a point on the surface of a sphere. A node is visible at each point. Each nodes radius is proportional to the probability $p_k$ of its basis state, whereas the node color indicates the quantum phase $\varphi_k$.

The nodes are laid out on the q-sphere so that the basis state with all zeros (e.g., $|000\rangle$) is at its north pole, and the basis state with all ones (e.g., $|111\rangle$) is at its south pole. Basis states with the same number of zeros (or ones) lie on a shared latitude of the q-sphere (e.g., $|001\rangle, |101\rangle$).

Beginning at the north pole of the q-sphere and progressing southward, each successive latitude has basis states with a greater number of ones; the latitude of a basis state is determined by its distance from the zero state. The q-sphere contains complete information about the quantum state in a compact representation.

<b>You can select, hold, and drag to rotate the q-sphere. To return the q-sphere to its default orientation, select the rewind-arrow button to the top right of the q-sphere.</b>

At this point, you have certainly asked yourself, Hank...isn't this just the Bloch sphere for multiple qubits?  Well IBM would very much like you to know it isn't

### <b>What is the difference between a Bloch sphere and a q-sphere?</b>

The q-sphere is not the Bloch sphere, even for a single qubit. Indeed, like the phase disk, the Bloch sphere gives a local view of the quantum state, where each qubit is viewed on its own. When trying to understand how multi-qubit states behave upon the application of quantum circuits, it is more insightful to take a global view. The q-sphere provides a visual representation of the quantum state.

If you have ever done programming before, you may have heard of a "Hello World" program, which is jargon for a simple program to introduce you to langauge.  Preparing and measuring a Bell state is what is often called the "Hello World" of quantum computing.

What is a Bell State?  Well its a state we have been talking about almost all school:

   $\LARGE|Bell\rangle=\frac{|00\rangle+|11\rangle}{\sqrt{2}}$

To prepare this 2-qubit state, you could proceed in two steps:

* First, prepare one of your qubits in the $\frac{|0\rangle+|1\rangle}{\sqrt{2}}$. To do this, you need to perform a Hadamard gate on this qubit.

* Next, you need to entangle two of your qubits through a two-qubit or "entangling gate".  In this case, a controlled-NOT gate: CNOT.  

When you have a controlled gate, it means that some qubits will be manipulated contingent upon the state of other ones.  For the CNOT on the composer, this is done by representing the <b>controlled</b> qubit by a (+) symbol, while the <b>controller</b> qubit is given by a (.)

* To prepare the Bell state, the qubit that was placed in the Hadamard state should be the controller.

You should now have a Bell state prepared.  <b>Do your Probablities and q-sphere correspond to your expectations?</b>

* Now, perform a measurement on each of the qubits.  You should see each gate snap to a connection to the classical qubits.

<b>How have the probabilities changed by including the measurements?</b>

You should see that we have "collapsed the wavefunction" by the measurement.  If you want to perform the quantum circuit again, but with a potentially different result, <b>change the number beside the "Visualization Seed".</b>

## Inspect

One final interesting feature of the Composer is "Inspect".  This feature is akin to the classical debugging feature where you are allowed to step through the function and see what condition the computer is in at any state. <i> In general, this is not possible for quantum computers because...measuring the state "collapses the wavefunction."</i> But since we are presently using a classical simulator of a small number of qubits, we can keep track of the full quantum state.
